```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
#from imblearn.metrics import classification_report_imbalanced
```

```python
companies_df = pd.read_csv('Unicorn_companies.csv')
new_companies_df = companies_df.drop(['Portfolio Exits'],axis = 1)
new_companies_df.head()
target = ['Valuation ($B)']
```

```python
new_companies_df['Valuation ($B)']= new_companies_df['Valuation ($B)'].str.replace('$','')
new_companies_df['Total Raised']= new_companies_df['Total Raised'].str.replace('$','')
new_companies_df['Total Raised']= new_companies_df['Total Raised'].str.replace('B','')

#new_companies_df['Valuation ($B)'] = new_companies_df['Valuation ($B)'].astype(string)
```

```python
lis = {}
len(new_companies_df["Industry"])
for x in range(0,len(new_companies_df["Industry"])):
    if new_companies_df["Industry"][x] not in lis:
        lis[new_companies_df["Industry"][x]] = 1
    else:
        lis[new_companies_df["Industry"][x]] += 1
    #print(new_companies_df["Industry"][x])
lis
```

```python
#df.columns = df.columns.str.replace('old_char', 'new_char')
new_companies_df.columns = new_companies_df.columns.str.replace('Total Raised', 'Total Raised ($B)')
```

```python
new_companies_df.head()
new_companies_df
```

```python
new_companies_df = new_companies_df.dropna(axis='columns', how='all')
```

```python
new_companies_df.head()
new_companies_df
```

```python
# Create our features
X = new_companies_df.drop('Valuation ($B)', axis=1)
X = pd.get_dummies(X)

# Create our target
y = new_companies_df.loc[:, target].copy()
```

```python
X.describe()
```

```python
# Check the balance of our target values
y['Valuation ($B)'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print(Counter(y_train['Valuation ($B)']))
print(Counter(y_test['Valuation ($B)']))
```

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
rf_model = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
rf_model.fit(X_train, y_train)
print(Counter(y_train['Valuation ($B)']))
```
